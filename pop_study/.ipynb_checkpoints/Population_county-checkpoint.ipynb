```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
from matplotlib.colors import LinearSegmentedColormap
```

```python
# Load the Texas counties shapefile
counties = gpd.read_file('County.shp')

# Load the population data
populations = pd.read_csv('Tx_population.csv')

# Merge the population data with the counties shapefile
merged_data = counties.merge(populations, right_on='CNTY_NM')
```

```python
# Define a custom colormap for the heatmap
colors = [(1, 1, 1), (0.8, 0.8, 1), (0.6, 0.6, 1), (0.4, 0.4, 1), (0.2, 0.2, 1), (0, 0, 0.8), (0, 0, 0.6), (0, 0, 0.4), (0, 0, 0.2), (0, 0, 0)]
cmap = LinearSegmentedColormap.from_list('my_colormap', colors)

# Plot the heatmap
fig, ax = plt.subplots(figsize=(12, 8))
merged_data.plot(column='Population', cmap=cmap, ax=ax)
ax.axis('off')
ax.set_title('Texas County Population Heatmap', fontdict={'fontsize': '25', 'fontweight' : '3'})

# Show the plot
plt.show()
```
